# Import Overlay
导入Overlay

```python
import pynq.lib.dma
import numpy as np

mmol = pynq.Overlay("matrixm.bit")

dma0 = mmol.axi_dma_0
dma1 = mmol.axi_dma_1
```

```python
#生成输入数据,并输出结果
from pynq import Xlnk
xlnk = Xlnk()
A = xlnk.cma_array(shape=(32,32), dtype=np.int)
B = xlnk.cma_array(shape=(32,32), dtype=np.int)
AB = xlnk.cma_array(shape=(32,32), dtype=np.int)

for i in range(32):
    for j in range(32):
        A[i][j] = 2;
        B[i][j] = 2;
               
dma0.sendchannel.transfer(A)
dma1.sendchannel.transfer(B)
dma0.recvchannel.transfer(AB)

print(A)
print(B)
print(AB)
```
